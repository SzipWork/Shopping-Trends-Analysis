# Descriptive analysis

```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

### To create a new categorical column ['Age_Category'] in the shop DataFrame by binning the values of the ['Age'] column into different age categories.

```python
shop['Age_Category'] = pd.cut(shop['Age'], bins = [0,15,18,30,50,70], labels = ['child','teen','young_adult','mid_age','old'])
```

```python
shop.head()
```

### 1. What is the overall distribution of customer ages in the dataset?

```python
fig = px.histogram(shop, x = 'Age', color = 'Age_Category')
fig.show()
```

```python
 import matplotlib.pyplot as plt
data = shop['Gender'].value_counts()
data.plot(kind='pie', explode=(0.1, 0.1), autopct='%1.1f%%')
plt.xlabel('Gender')
plt.show()
```

### 2.How does the average purchase amount vary across different product categories?

```python
import matplotlib.pyplot as plt
ax = shop['Category'].value_counts().plot(kind='bar', rot=0)
for p in ax.patches:
    ax.annotate(int(p.get_height()),
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center')
plt.xlabel("Category", weight='bold')
plt.ylabel("Number of Occurrence", weight='bold')


plt.show()
```

  ### 3.Which gender has the highest number of purchases?
